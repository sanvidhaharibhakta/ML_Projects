```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#feature selection
selected_features= ['keywords','genres','tagline','cast','director']
```

```python
selected_features
```

```python
#Preprocessing
for feature in selected_features:
  movies_data[feature]= movies_data[feature].fillna('')
```

```python
Combine_features = movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['cast']+''+movies_data['director']
```

```python
Combine_features
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(Combine_features)
```

```python
#Similarity Confidence Value

similarity = cosine_similarity(feature_vectors)
```

MOVIE RECOMMENDATION SYSTEM

```python
movie_name = input('Enter your Favourite Movie Name :')

title_list = movies_data['title'].tolist()

Close_match= difflib.get_close_matches(movie_name, title_list)

close_match = Close_match[0]

index_m = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_m]))

sorted_similar_score = sorted(similarity_score, key= lambda x:x[1], reverse= True)

print('Movies Suggested According to your Preference: \n')

i = 1

for movie in sorted_similar_score:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i += 1
```
