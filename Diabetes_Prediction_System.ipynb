```python
#Importing Dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#Data collection
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
#Statistical measures
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts
# 0 = Non diabetic
# 1 = Diabetic
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#Feature Selection
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
#Data Standardisation

scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
X= standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify = Y, random_state = 2)
```

Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
pred = classifier.predict(X_train)
```

```python
acc = accuracy_score(pred, Y_train)
```

```python
print("Accuracy score of training data: ", acc)
```

```python
xtest_pred = classifier.predict(X_test)
acc_test = accuracy_score(xtest_pred, Y_test)
print("Accuracy score of test data: ", acc_test)
```

**PREDICTIVE SYSTEM**

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

input_data_as_array = np.asarray(input_data)

input_reshaped = input_data_as_array.reshape(1, -1)

std_data= scaler.transform(input_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction [0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
